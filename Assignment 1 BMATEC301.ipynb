```python
import sympy
from sympy import *
init_printing()
x = Symbol('x')
y = Function('y')(x)
t = Symbol('t')
x = Function('x')(t)
```

### Question 1
### Solve $\frac{d^4y}{dx^4} + \frac{d^3y}{dx^3} + \frac{d^2y}{dx^2} = 5x^2 + cos(x)$

```python
eqn = Eq(y.diff('x', 4) + y.diff('x', 3) + y.diff('x', 2) - 5*x**2 - cos(x) , 0)
solution = dsolve(eqn)
# pprint(solution, use_unicode=True)
solution
```

### Question 2
### The following differential equation is the equation of motion for an ideal spring-mass system with damping and an external force $F(t)$.
### $$m\frac{d^2x}{dt^2} + c\frac{dx}{dt} + kx = F(t) $$
### If $m=2kg$, $c=4Ns\cdot{m^{-1}}$, $k=6Nm^{-1}$, $F=7N$ and $x(0)=x'(0)=0$, find the solution of the differential equation.

```python
eqn = Eq((x.diff(t, 2))*2 + 6*x.diff(t) + 6*x - 7, 0)
solution = dsolve(eqn)
solution
```

### Question 3
### Solve $\frac{d^2y}{dx^2} - 2\frac{dy}{dx} + y = e^x\cdot{log(x)}$ by method of variation of parameters.

```python
eqn = Eq(y.diff(x, 2) - 2*y.diff(x) + y - exp(x)*log(x), 0)
solution = dsolve(eqn)
solution
```
