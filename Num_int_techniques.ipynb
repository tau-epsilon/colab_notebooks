## Problem 1
### 1.) Evaluate  $~\large \int_{0}^{100} \frac{x \cdot sin(x)} {1 + \tan^{-1}(x)}$ using midpoint formula.

## Solution

### Plotting the graph of the function using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
# Lambda function for integrand
f  = lambda y:y*np.sin(y)/(1 + np.arctan(y))
# Generating x values
x = np.linspace(0,100,10000)
# Corresponding function values
y = f(x)
# Size of the figure
plt.figure(figsize=(10,5))
# Title of the figure
plt.title(r"Plot of $ \frac{xsin(x)} {1 + \tan^{-1}(x)}~versus~x $", fontsize=16)
# Label for x-axis
plt.xlabel("x", fontsize=15)
# Label for y-axis
plt.ylabel("f(x)", fontsize=15)

# Limits for x and y axes

plt.grid(True, which='both', linestyle='-', linewidth=0.9)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.xlim(0, 100)
plt.ylim(np.min(y), np.max(y))


# Center the axes
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['left'].set_color('black')
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_position('zero')
ax.spines['bottom'].set_color('black')
ax.spines['bottom'].set_linewidth(1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#Plotting the relationship
plt.plot(x,y,
         linewidth=5, # Width of lines in line plot
         label=r'$f(x) = \frac{x \sin(x)}{1 + \tan^{-1}(x)}$' # Label for adding legends
        )
plt.fill_between(x, y, color='#5FEAB4', alpha=0.9)
plt.legend(fontsize=12)
plt.xticks([])
plt.show()
```

### Actual integration

```python
import numpy as np

f = lambda y: y*np.sin(y)/(y + np.tan(y))

def midpoint_integral(start, stop, steps):
    integral = 0
    step = (stop - start)/steps
    for i in range(steps):
        integral += f(start + step/2)*step
        start += step
    return integral
start = 0
stop = 100
steps = 100
integral_complete = midpoint_integral(start, stop, steps)
print(f"Ans -- {integral_complete}")
```

```python
plt.style.use('fivethirtyeight')
x = [i for i in range(1,500)]
y = [midpoint_integral(0, 100, i) for i in x]
plt.plot(x,y, marker="o", linewidth=1)
plt.show()
```
