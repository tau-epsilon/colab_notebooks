## Root of an equation


### Plotting function graph

A number $\zeta$ is a root of an equation $f(x)=0$ if $f(\zeta) = 0$ and $~f(x)\mod(x-a) = 0$

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-100, 100, 10000)
y = x*np.sin(x) - np.arctan(x)
plt.figure(figsize=(15,10))
plt.axhline(y=0, color='r', linestyle='--', label="y=0")
plt.axvline(x=0, color='r', linestyle='-', label="x=0")
plt.xticks(range(-100,100,10), labels=[i for i in range(20)])
plt.yticks(range(-100, 100, 10))
plt.plot(x,y, label="Plot of y vs x")
plt.savefig('A plot')
plt.legend()
plt.fill(x,y,'b', alpha=0.12)
# plt.savefig('A_plot',transparent=12, dpi=720, edgecolor='r')
plt.axis(False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = np.linspace(0, 10, 100)
y = 4 + 1 * np.sin(2 * x)
x2 = np.linspace(0, 10, 25)
y2 = 4 + 1 * np.sin(2 * x2)

# plot
fig, ax = plt.subplots()

ax.plot(x2, y2 + 2.5, 'x', markeredgewidth=2)
ax.plot(x, y, linewidth=2.0)
ax.plot(x2, y2 - 2.5, 'o-', linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
