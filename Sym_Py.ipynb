```python
import sympy as sp

# Define symbols for variables
x, y, z = sp.symbols('x y z')

# Define the components of the vector field
F_x = x + y + 1
F_y = 1
F_z = -(x+y)

# Compute the gradients
grad_F_x = sp.diff(F_x, x)
grad_F_y = sp.diff(F_y, y)
grad_F_z = sp.diff(F_z, z)

# Display the gradients
print("Gradient of F_x:", grad_F_x)
print("Gradient of F_y:", grad_F_y)
print("Gradient of F_z:", grad_F_z)
```

```python
import sympy as sp

# Define symbols for variables
x, y, z = sp.symbols('x y z')

# Define the components of the vector field
F_x = x + y + 1
F_y = 1
F_z = -(x + y)

Vec_Mat = sp.Matrix([
    F_x,
    F_y,
    F_z
])
# Compute the curl
curl_F = sp.Matrix([
    sp.diff(F_z, y) - sp.diff(F_y, z),
    sp.diff(F_x, z) - sp.diff(F_z, x),
    sp.diff(F_y, x) - sp.diff(F_x, y)
])

curl_F.dot(Vec_Mat)
```

 **1. Find directional derivative of scalar valued function,
 $f(x,y,z)=\frac{xz}{x^2 + y^2}$ at a point $(1 ,-1, 1)$
 along the vector $\hat{i} - 2\hat{j} + 3\hat{k}$**


*Solution :*

**Steps**


*   Find $∇⋅f$
*   Substitute $(1,-1,1)$ in place of $(x,y,z)$
*   Find $\hat{a}$ by taking $|\hat{i} - 2\hat{j} + 3\hat{k}|$ and dividing by $\vec{a}$
*   Then finding $DD$ by solving $∇f·\hat{a}$




```python
# Finding the grad(F) (vector function)
x, y, z = symbols('x,y,z') # Initialized symbols x, y, z

scalar_function = ( (x * z) / (x**2 + y**2))
Grad_Matrix = Matrix([
    simplify(diff(scalar_function, x)),
    simplify(diff(scalar_function, y)),
    simplify(diff(scalar_function, z))
])

display(Grad_Matrix) # Printing Gradient of f

```

```python
substituted_Grad_Matrix = Grad_Matrix.subs({x: 1, y: -1, z: 1}) # Substitution.

display(substituted_Grad_Matrix)
```

```python
# Now taking modulus of a vector

a_vec = Matrix([[
    1,-2,3
]])

unit_vec = a_vec/sqrt(6)

unit_vec
```

```python
unit_vec.dot(substituted_Grad_Matrix)
```
